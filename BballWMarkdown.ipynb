```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline

baturl = 'https://raw.githubusercontent.com/danielmoore19/baseball/master/baseballdatabank-2019.2/core/Batting.csv'
piturl = 'https://raw.githubusercontent.com/danielmoore19/baseball/master/baseballdatabank-2019.2/core/Pitching.csv'

batdata = pd.read_csv(baturl)

batdata.describe()
```

```python
lowerMound = batdata[(batdata['yearID'] >= 1968) & (batdata['yearID'] <= 1991) & (batdata['AB'] >= 502)]
steroidEra = batdata[(batdata['yearID'] >= 1991) & (batdata['yearID'] <= 2003) & (batdata['AB'] >= 502)]
statsEra = batdata[(batdata['yearID'] >= 2003) & (batdata['AB'] >= 502)]
```

```python
print(lowerMound.describe())
sns.regplot(x='H', y='HR', data=lowerMound)
plt.show()
```

```python
print(steroidEra.describe())
sns.regplot(x='H', y='HR', data=steroidEra)
plt.show()
```

```python
print(statsEra.describe())
sns.regplot(x='H', y='HR', data=statsEra)
plt.show()
```

```python
last5 = batdata[(batdata['yearID'] >= 2013) & (batdata['AB'] >= 502)]
print(last5.describe())
sns.regplot(x='H', y='HR', data=last5)
plt.show()
```

Based on qualified hitters (3.1 at bats per team game), the statsEra shows a higher homerun (HR) median than when MLB lowered the mound in 1968, but a lower median than the steroidEra. This holds on the median of hits (H) and runs scored (R) across the eras. At this points these stats have not been measured against pitching over the same eras.

Two things that stand out against the other data is the decline of the stolen base (SB), and the increase in doubles (2B). Stats have effectively killed the SB requiring such a high success rate that teams simply do not "run", where run means attempt to steal bases. However, the increased stats regarding optimal launch angles has created an increase in 2B during the statEra that is higher than both other eras.

That said, and even though MLB has set records for R and HR in each of the last three seasons, it has not yielded an increase in the median for R, 2B, or H from qualified players. There is a small increase (.08) in HR.

The importance of Qualified Players is these players make up the lionshare of money devoted to players by teams. No one pays a part time player 10m a season, but they will pay an every day player 30mil. If front offices see that their money is creating negligible increases or regression then it will change the way they spend resources on players. This is why it necessary to also measure pitching. As hitters have moved to statistics based approaches at the plate, it creates a need for ptichers to respond. As they respond, so in kind will then hitters.
